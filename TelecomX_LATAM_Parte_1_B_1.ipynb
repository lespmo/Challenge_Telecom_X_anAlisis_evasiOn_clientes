```python
import pandas as pd
# funcion de informe unquie y tip de datos
def reporte_unique(df):
    reporte = {
        "columna": [],
        "tipo_dato": [],
        "n_valores_unicos": [],
        "valores": []
    }

    for col in df.columns:
        reporte["columna"].append(col)
        reporte["tipo_dato"].append(df[col].dtype)
        reporte["n_valores_unicos"].append(df[col].nunique())
        reporte["valores"].append(df[col].unique())

    return pd.DataFrame(reporte)

```
